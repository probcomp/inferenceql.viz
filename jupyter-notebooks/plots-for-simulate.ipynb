```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


%matplotlib inline
```

```python
joint = pd.read_csv('../results/csv-files/data.csv')
```

```python
lims= (0, 18)
```

```python
g = sns.pairplot(joint, vars=['x', 'y'], hue='z')
g.set(xlim=lims, ylim=lims)
g.fig.set_size_inches(10, 10)
```

```python
samples = pd.read_csv('../results/csv-files/x-y-given-z-eq-0.csv')
fig, ax = plt.subplots()
ax.scatter(samples['x'], samples['y'], color='b', s=2)
ax.set_xlim(lims)
ax.set_ylim(lims)
ax.set_xlabel('x')
ax.set_ylabel('y')
fig.set_size_inches(6, 5.5)
```
